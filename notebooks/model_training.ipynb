```python

from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# PyTorch
import torch
print(f"GPU available: {torch.backends.mps.is_available()}")

# TensorFlow
import tensorflow as tf
print(f"GPUs: {tf.config.list_physical_devices('GPU')}")

# Define augmentations
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load from directory structure (train/class_name/images)
train_dataset = datasets.ImageFolder('/Users/tawate/Documents/Architecture_Image_Modeling/architectural_style_images/train', transform=train_transforms)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2]
)

train_generator = train_datagen.flow_from_directory(
    '/Users/tawate/Documents/Architecture_Image_Modeling/architectural_style_images/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

```
